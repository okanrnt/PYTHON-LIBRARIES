```python
import pandas as pd
```

```python
#  the top 10 countries with the highest life expectancies

women = {
    'Hong Kong':   87.6,
    'Japan':       87.5,
    'Switzerland': 85.5,
    'Singapore':   85.7,
    'Italy':       85.4,
    'Spain':       86.1,
    'Australia':   85.3,
    'Iceland':     84.4,
    'Israel':      84.4,
    'South Korea': 85.8
}

men = {
    'Hong Kong':   81.8,
    'Japan':       81.3,
    'Switzerland': 81.7,
    'Singapore':   81.4,
    'Italy':       81.1,
    'Spain':       80.7,
    'Australia':   81.3,
    'Iceland':     81.3,
    'Israel':      81.1,
    'South Korea': 79.7
}
```

## 1- Write a DataFrame with two columns Women and Men

```python
life_exp_top10 = pd.DataFrame({"Women":women,
                              "Men":men})
life_exp_top10
```

## 2- Add a new column to life_exp_top10 named Average

```python
Average = life_exp_top10.mean(axis=1)
life_exp_top10["Average"] = Average
life_exp_top10
```

## 3- Add the provided rank Series as a new column of life_exp_top10 named Rank

```python
rank = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9, 9])

print(len(life_exp_top10.index))
print(len(rank.index))
rank.index = life_exp_top10.index
life_exp_top10["Rank"] = rank
life_exp_top10
```

## 4- Insert the provided rank Series as the first column of life_exp_top10 named Rank

```python
life_exp_top10.drop('Rank',axis=1,inplace=True)

rank = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9, 9])
rank.index = life_exp_top10.index
life_exp_top10.insert(0, 'Rank', rank)
life_exp_top10
```

## 5- Add a new row for Sweden

```python
life_exp_top10.drop('Rank',axis=1,inplace=True)
dataa = {"Women":84.4,"Men":80.9,"Average":(84.4 + 80.9) / 2}
life_exp_top10.loc["Sweden"] = dataa
life_exp_top10
```

## 6- Remove the row containing Italy's life expectancy data

```python
life_exp_top10 = life_exp_top10.drop("Italy")
life_exp_top10
```

```python
life_exp_top10 = life_exp_top10.reindex(columns= ["Average","Women","Men"])
life_exp_top10
```
