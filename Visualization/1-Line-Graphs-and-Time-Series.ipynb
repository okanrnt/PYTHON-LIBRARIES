## 3. Time Series Data

```python
who_time_series = pd.read_csv('covid-vaccination-vs-death_ratio.csv')
```

```python
who_time_series.drop(['Unnamed: 0','iso_code'],axis=1,inplace=True)
```

```python
who_time_series['date'] = pd.to_datetime(who_time_series['date'])
```

## 4. Defining a Function for Visualization

```python
def plot_vaccinated_people(country_name):
    country = who_time_series[who_time_series['country'] == country_name]
    plt.plot(country['date'], country['people_vaccinated'])
    plt.title('{}: Vaccinated People'.format(country_name))
    plt.xlabel('Date')
    plt.ylabel('Number of Vaccinated')
    plt.show()
plot_vaccinated_people('France')
plot_vaccinated_people('Iceland')
plot_vaccinated_people('Argentina')
```

## 5. Comparing Line Graphs

```python
Turkey = who_time_series[who_time_series['country'] == 'Turkey']
Spain = who_time_series[who_time_series['country'] == 'Spain']
Italy = who_time_series[who_time_series['country'] == 'Italy']
plt.plot(Turkey['date'],Turkey['people_vaccinated'],label='Turkey')
plt.plot(Spain['date'],Spain['people_vaccinated'],label='Spain')
plt.plot(Italy['date'],Italy['people_vaccinated'],label='Italy')
plt.legend()
plt.show()
```
