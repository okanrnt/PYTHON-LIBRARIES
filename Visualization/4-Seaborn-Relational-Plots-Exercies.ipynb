```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Introduction

```python
housing = pd.read_csv('miami-housing.csv')
```

```python
housing.head()
```

```python
housing['SALE_PRC'] = housing['SALE_PRC'].astype(int)
```

```python
housing.tail()
```

```python
housing.info()
```

## 2. Seaborn

```python
sns.set_theme()
sns.relplot(data=housing,x='LND_SQFOOT',y='SALE_PRC')
plt.show()
```

## 3. Variable Representation: Color

```python
sns.set_theme()
sns.relplot(data=housing,x='LND_SQFOOT',y='SALE_PRC',hue='structure_quality',palette='RdYlGn')
plt.show()
```

## 4. Variable Representation: Size

```python
sns.set_theme()
sns.relplot(data=housing,x='LND_SQFOOT',y='SALE_PRC',hue='structure_quality',palette='RdYlGn',size='TOT_LVG_AREA',sizes=(1,300))
plt.show()
```

## 5. Visualitizion Older and Newer Houses

```python
over_50_ages = housing[housing['age'] >= 50]
below_50_ages = housing[housing['age'] < 50]
```

```python
sns.set_theme()
sns.relplot(data=below_50_ages,x='LND_SQFOOT',y='SALE_PRC',hue='structure_quality',palette='RdYlGn',
            size='TOT_LVG_AREA',sizes=(1,300))
plt.show()
```

```python
sns.set_theme()
sns.relplot(data=over_50_ages,x='LND_SQFOOT',y='SALE_PRC',hue='structure_quality',palette='RdYlGn',
            size='TOT_LVG_AREA',sizes=(1,300))
plt.show()
```
